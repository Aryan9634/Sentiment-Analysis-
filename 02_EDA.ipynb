**02 : EDA**

<u>Importing required modules</u>

```python
#Data pre-processing 
import numpy as np
import pandas as pd

#Data visualisation 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#Handling warnings
import warnings
warnings.filterwarnings("ignore")
```

<u>Loading data</u>

```python
data = pd.read_csv("processed_train")
data.head(3)
```

<u>Data visualisation</u>

```python
sns.set_style("dark")
```

```python
cols = ["sentiment","tweet_time","user_age"]
plt.figure(figsize=(15,5))
for i in range(1,len(cols)+1):
    plt.subplot(1,3,i)
    sns.countplot(data[cols[i-1]])
    plt.title(cols[i-1].upper())
    plt.xlabel(None)
    plt.ylabel(None)
    plt.grid()
```

*Inferences;*
- Maximum number of tweets are neutral followed by positive and then negative.
- Almost same number of tweets are recieved irrespective of times of day.
- All age groups tweet similar number of times

```python
plt.figure(figsize=(16,5))
x = data.Country.value_counts().index[65:130]
y = data.Country.value_counts().values[65:130]
plt.plot(x,y,marker="o",ms=2.5,color="r")
plt.xticks(rotation=90)
plt.grid(linewidth=0.5,color="g")

x = np.linspace(0,64)
y = [138]*len(x)
plt.plot(x,y,)
plt.xlim(0,64)

plt.show()
```

*Inferences*
- Above graph represents the flucatuation of quantity of tweets wrt to countries

```python
plt.figure(figsize=(16,5))

ax1 = plt.subplot2grid((1,3),(0,0),colspan=2)
ax1 = sns.countplot(data.user_age,hue=data.sentiment)
plt.title("User age vs sentiment")
plt.xlabel(None)
plt.ylabel(None)
plt.grid()

ax2 = plt.subplot2grid((1,3),(0,2),colspan=1)
ax2 = sns.countplot(data.tweet_time,hue=data.sentiment)
plt.title("Tweet time vs sentiment")
plt.xlabel(None)
plt.ylabel(None)
plt.grid()

plt.show()
```

*Inferences;*
- Ratio of tweets stays constant wrt to user ages, where neutral tweets exceed positive tweets further followed by negative tweets
- Ratio of tweets stays constant wrt to times of day, where neutral tweets exceed positive tweets further followed by negative tweets

```python
plt.figure(figsize=(16,5))
flierprops = dict(marker='o', markersize=2.5, markeredgecolor='black', alpha=0.2)

plt.subplot(1,3,1)
plt.boxplot(data.population,flierprops=flierprops)
plt.ylim(-0.1*1e8,0.1*1e9)
plt.title("POPULATION")
plt.grid()

plt.subplot(1,3,2)
plt.boxplot(data.land_area,flierprops=flierprops)
plt.ylim(-0.1*1e6,0.2*1e7)
plt.title("LAND AREA")
plt.grid()

plt.subplot(1,3,3)
plt.boxplot(data.density,flierprops=flierprops)
plt.ylim(-0.5,700)
plt.title("DENSITY")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16,18))
plt.subplot(3,1,1)
plt.plot(data.Country,data.population,color="r",linewidth=0.7)
plt.title("POPULATION")
plt.xticks(rotation=90)
plt.xlim(30,80)
plt.grid(linewidth=0.5,color="g")

plt.subplot(3,1,2)
plt.plot(data.Country,data.land_area,color="r",linewidth=0.7)
plt.title("LAND AREA")
plt.xticks(rotation=90)
plt.xlim(130,180)
plt.grid(linewidth=0.5,color="g")

plt.subplot(3,1,3)
plt.plot(data.Country,data.density,color="r",linewidth=0.7)
plt.title("DENSITY")
plt.xticks(rotation=90)
plt.xlim(110,160)
plt.grid(linewidth=0.5,color="g")

plt.tight_layout()
plt.show()
```

*Inferences;*
- China has the maximum population followed by India.
- Russia has the maximum land area.
- Monaco has the maximum density followed by Singapore
